```python
import pandas as pd
import warnings
import seaborn as sns
warnings.filterwarnings("ignore")

colspecs = [
    (0, 21),
    (21, 43),
    (43, 49),
    (51, 56),
    (58, 65),
    (67, 75),
    (77, 85)
]

column_names = [
    "Airport",
    "City",
    "Scheduled departures",
    "Performed departures",
    "Enplaned passengers",
    "Enplaned revenue tons of freight",
    "Enplaned revenue tons of mail"
]

df = pd.read_fwf("/kaggle/input/airport/airport.dat.txt", colspecs=colspecs, header=None, names=column_names, 
                    dtype={
                        "Scheduled departures": int,
                        "Performed departures": int,
                        "Enplaned passengers": int,
                        "Enplaned revenue tons of freight": float,
                        "Enplaned revenue tons of mail": float
                    })
```

```python
cols = ['Scheduled departures', 'Performed departures', 
            'Enplaned passengers', 'Enplaned revenue tons of freight', 
            'Enplaned revenue tons of mail']
for col in cols:
    df[col] = pd.to_numeric(df[col].astype(str).str.strip(), errors='coerce')
```

```python
descriptive_stats = df[cols].describe()
descriptive_stats
```

```python
mean = (descriptive_stats.loc['mean']
variance = (descriptive_stats.loc['std'] ** 2).round(3)
std_dev = descriptive_stats.loc['std'].round(3)
median = ((descriptive_stats.loc['max'] - descriptive_stats.loc['min']) / 2).round(3)
first_quartile = descriptive_stats.loc['25%'].round(3)
third_quartile = descriptive_stats.loc['75%'].round(3)

# Ð’Ñ‹Ð²Ð¾Ð´ Ñ€ÐµÐ·ÑƒÐ»ÑŒÑ‚Ð°Ñ‚Ð¾Ð²
print("Ð’Ñ‹Ð±Ð¾Ñ€Ð¾Ñ‡Ð½Ð¾Ðµ ÑÑ€ÐµÐ´Ð½ÐµÐµ:\n", mean)
print("\nÐ’Ñ‹Ð±Ð¾Ñ€Ð¾Ñ‡Ð½Ð°Ñ Ð´Ð¸ÑÐ¿ÐµÑ€ÑÐ¸Ñ:\n", variance)
print("\nÐ¡Ñ‚Ð°Ð½Ð´Ð°Ñ€Ñ‚Ð½Ð¾Ðµ Ð¾Ñ‚ÐºÐ»Ð¾Ð½ÐµÐ½Ð¸Ðµ:\n", std_dev)
print("\nÐœÐµÐ´Ð¸Ð°Ð½Ð°:\n", median)
print("\n1-Ð¹ ÐºÐ²Ð°Ñ€Ñ‚Ð¸Ð»ÑŒ:\n", first_quartile)
print("\n3-Ð¹ ÐºÐ²Ð°Ñ€Ñ‚Ð¸Ð»ÑŒ:\n", third_quartile)
```
